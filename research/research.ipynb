```python
## Esnure annotation

def get_product(x: int, y: int) -> int:
    return x * y
```

```python
get_product(2, 3)
```

```python
get_product(2, "3")
```

```python
from ensure import ensure_annotations

@ensure_annotations
def get_product(x: int, y: int) -> int:
    return x * y

get_product(2, "3")
```
